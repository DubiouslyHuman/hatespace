```python
import torch
import numpy as np
import pandas as pd
import seaborn
```

```python
embedding_data = pd.read_csv('saved_models\\vae\sinkhorn\\alpha_0.2_latent_16_softmax_True_version_1.0.0_features_True')
embedding_data = embedding_data.to_numpy()


embedding_distances = np.power(
        np.sum(
            np.power(
                np.abs(embedding_data - (1 / embedding_data.shape[1]))
            , 2.0)
        , axis = 1)
    , 0.5)
```

```python
dirichlet_distribution = torch.distributions.dirichlet.Dirichlet(torch.tensor([0.2 for i in range(embedding_data.shape[1])]))
dirichlet_data = dirichlet_distribution.sample([1000000]).numpy()

dirichlet_distances = np.power(
        np.sum(
            np.power(
                np.abs(dirichlet_data - (1 / embedding_data.shape[1]))
            , 2.0)
        , axis = 1)
    , 0.5)
```

```python
seaborn.displot(dirichlet_distances, kind = 'kde', bw_adjust = 0.5)
```

```python
seaborn.displot(embedding_data)
```

```python
seaborn.displot(embedding_distances, kind = 'kde', bw_adjust = 0.5)
```
